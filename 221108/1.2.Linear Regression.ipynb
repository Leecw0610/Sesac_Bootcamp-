# Boston House Price

```python
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split #train / test 나누기
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
dataset = sklearn.datasets.load_boston()
```

```python
data= pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
data['price']=dataset.target
```

```python
data.head()
```

```python
model = LinearRegression()
```

```python
x = data.drop('price',axis=1)
```

```python
y= data['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
model.fit(x_train, y_train)
```

```python
pre_train = model.predict(x_train)
```

```python
mean_absolute_error(pre_train, y_train) #적을수록 좋음
```

```python
model.score(x_train, y_train) #모델 정확도
```

```python
pre_test = model.predict(x_test)
```

```python
mean_absolute_error(pre_test, y_test)
```
